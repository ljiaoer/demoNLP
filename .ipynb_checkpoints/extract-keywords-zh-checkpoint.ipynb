```python
import jieba
from jieba.analyse import *
```

```python
with open ("sample-zh.txt",encoding="utf-8") as f:
    data = f.read()
```

```python
data
```

```python
# 使用TF-IDF提取关键词和权重，并以此显示出来,默认为20个
for keyword,weight in extract_tags(data,withWeight=True):
    print("%s %s" %(keyword,weight))
```

```python
# 使用TF-IDF提取关键词和权重，并显示10个
for keyword,weight in extract_tags(data,withWeight=True,topK=10):
    print("%s %s" %(keyword,weight))
```

```python
# 使用TextRank提取关键词
for keyword,weight in textrank(data, withWeight=True):
    print("%s %s" %(keyword,weight))
```
