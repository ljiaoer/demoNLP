```python
import pandas as pd
import re
```

```python
with open ("story.txt") as f:
    data = f.read()
```

```python
print(data)
```

```python
# # 去除开头部分不相关内容
data = data.split("[Opening Credits]")[1]
```

```python
print(data)
```

```python
# 去除结尾部分不相关内容
data = data.split("[End Credits")[0]
```

```python
print(data)
```

```python
# /去除空行
regex = "^$\n"
subst = ""
data = re.sub(regex,subst,data,0,re.MULTILINE)
```

```python
print(data)
```

```python
# 去除分割线
regex = r"^-+$\n"
subst = ""
data = re.sub(regex, subst, data, 0, re.MULTILINE)
```

```python
print(data)
```

```python
# 利用换行符把原本完整的文本分割成行。
lines = data.split("\n")
```

```python
print(lines)
```

```python
# 然后给每一行加上行号。
myrows = []
num = 1
for line in lines:
    myrows.append([num,line])
    num = num + 1
```

```python
print(myrows)
```

```python
myrows[:3]
```

```python
myrows[3:]
```

```python
# 数组转换成数据框
df = pd.DataFrame(myrows)
```

```python
df.head()
```

```python
df.columns = ["lines","text"]
```

```python
df.head()
```

```python
df.to_csv("story.csv",index = False)
```
